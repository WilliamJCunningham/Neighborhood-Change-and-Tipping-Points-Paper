```python
def get_input_ids_and_attention_mask_chunk():
    """
    This function splits the input_ids and attention_mask into chunks of size 'chunksize'. 
    It also adds special tokens (101 for [CLS] and 102 for [SEP]) at the start and end of each chunk.
    If the length of a chunk is less than 'chunksize', it pads the chunk with zeros at the end.
    
    Returns:
        input_id_chunks (List[torch.Tensor]): List of chunked input_ids.
        attention_mask_chunks (List[torch.Tensor]): List of chunked attention_masks.
    """
    chunksize = 512
    input_id_chunks = list(tokens['input_ids'][0].split(chunksize - 2))
    attention_mask_chunks = list(tokens['attention_mask'][0].split(chunksize - 2))
    
    for i in range(len(input_id_chunks)):
        input_id_chunks[i] = torch.cat([
            torch.tensor([101]), input_id_chunks[i], torch.tensor([102])
        ])
        
        attention_mask_chunks[i] = torch.cat([
            torch.tensor([1]), attention_mask_chunks[i], torch.tensor([1])
        ])
        
        pad_length = chunksize - input_id_chunks[i].shape[0]
        
        if pad_length > 0:
            input_id_chunks[i] = torch.cat([
                input_id_chunks[i], torch.Tensor([0] * pad_length)
            ])
            attention_mask_chunks[i] = torch.cat([
                attention_mask_chunks[i], torch.Tensor([0] * pad_length)
            ])
            
    return input_id_chunks, attention_mask_chunks 
```

```python
import pandas as pd
import numpy as np

def get_gender_coefficient(coefficients):
    return next((v for k, v in coefficients.items() if 'gender' in k.lower()), np.nan)

def get_gender_pvalue(pvalues):
    return next((v for k, v in pvalues.items() if 'gender' in k.lower()), np.nan)

def add_stars(coefficient, p_value):
    if pd.isna(coefficient) or pd.isna(p_value):
        return str(coefficient)
    stars = ''
    if p_value < 0.01:
        stars = '***'
    elif p_value < 0.05:
        stars = '**'
    elif p_value < 0.1:
        stars = '*'
    return f"{coefficient:.4f}{stars}"

def create_structured_table(results_df):
    author_results = results_df[results_df['dependent_var'].str.contains('To_Author')]
    editor_results = results_df[results_df['dependent_var'].str.contains('To_Editor')]

    structured_results = []

    for author_var in author_results['dependent_var']:
        author_row = author_results[author_results['dependent_var'] == author_var].iloc[0]
        editor_var = author_var.replace('To_Author', 'To_Editor')
        editor_row = editor_results[editor_results['dependent_var'] == editor_var].iloc[0]
        
        author_coef = get_gender_coefficient(author_row['coefficients'])
        author_pvalue = get_gender_pvalue(author_row['p_values'])
        editor_coef = get_gender_coefficient(editor_row['coefficients'])
        editor_pvalue = get_gender_pvalue(editor_row['p_values'])

        combined_row = {
            'Sentiment Measure': author_var.split('_To_Author_')[-1],
            'Model': author_var.split('_To_')[0],
            'To Author': add_stars(author_coef, author_pvalue),
            'To Editor': add_stars(editor_coef, editor_pvalue)
        }
        structured_results.append(combined_row)

    return pd.DataFrame(structured_results)

# Create the structured table
structured_table = create_structured_table(results_df)

# Display the table
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
print(structured_table.to_string(index=False))

# Export to CSV
structured_table.to_csv('structured_regression_results.csv', index=False)
print("\nStructured regression results exported to 'structured_regression_results.csv'.")
```
